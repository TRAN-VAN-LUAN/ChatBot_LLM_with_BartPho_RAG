```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv('data/csv/articles.csv')

# Kiểm tra số lượng giá trị bị thiếu trong mỗi cột trước khi drop
print("Missing values before drop:")
print(df.isnull().sum())

# Xóa tất cả các hàng chứa giá trị null và cập nhật lại index
df_cleaned = df.dropna().reset_index(drop=True)

# In ra DataFrame sau khi xóa các hàng có giá trị null và cập nhật lại index
print("DataFrame cleaned and reindexed:")
df_cleaned
```

```python
df_cleaned.to_csv('data/csv/cleaned_articles.csv', index=False)
```
