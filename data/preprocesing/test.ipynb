```python
import pandas as pd
```

```python
df = pd.read_csv('./tamanh_hosipital_dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
# Giả sử df là DataFrame của bạn
# Kiểm tra số lượng giá trị trùng lặp trong cột 'Title'
duplicates_count_title = df['Question'].duplicated().sum()
duplicates_count_content = df['Answer'].duplicated().sum()
print(f"Số lượng giá trị trùng lặp trong cột 'Question': {duplicates_count_title}")
print(f"Số lượng giá trị trùng lặp trong cột 'Answer': {duplicates_count_content}")

```

```python
df = df.drop_duplicates(subset=['Question'])
df =  df.dropna(axis=1, how='all')
```

```python
df = df.drop_duplicates()
```

```python
# Giả sử df là DataFrame của bạn
# Kiểm tra số lượng giá trị trùng lặp trong cột 'Title'
duplicates_count_title = df['Question'].duplicated().sum()
duplicates_count_content = df['Answer'].duplicated().sum()
print(f"Số lượng giá trị trùng lặp trong cột 'Question': {duplicates_count_title}")
print(f"Số lượng giá trị trùng lặp trong cột 'Answer': {duplicates_count_content}")

```

```python
# Lọc các tiêu đề chứa 'Mục lục'
mask = df['Question'].str.contains('Mục lục', case=False, na=False)

# Xóa các hàng chứa tiêu đề với 'Mục lục'
df = df[~mask]
df = df.reset_index(drop=True)

# In ra kết quả
df
```

```python
# Lọc các tiêu đề chứa 'Mục lục'
mask = df['Question'].str.contains('HỆ THỐNG BỆNH VIỆN ĐA KHOA TÂM ANH', case=False, na=False)

# Xóa các hàng chứa tiêu đề với 'Mục lục'
df = df[~mask]
df = df.reset_index(drop=True)

# In ra kết quả
df
```

```python
# Lưu DataFrame vào file CSV
df.to_csv('../csv/processed_tamanh_hospital_cleaned.csv', index=False, encoding='utf-8')
```
